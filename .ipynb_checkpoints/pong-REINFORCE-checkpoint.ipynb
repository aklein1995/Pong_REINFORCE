# Policy

## Exercise 1: Implement your policy
 
Here, we define our policy. The input is the stack of two different frames (which captures the movement), and the output is a number $P_{\rm right}$, the probability of moving left. Note that $P_{\rm left}= 1-P_{\rm right}$

```python
import torch
import torch.nn as nn
import torch.nn.functional as F



# set up a convolutional neural net
# the output is the probability of moving right
# P(left) = 1-P(right)
class Policy(nn.Module):

    def __init__(self):
        super(Policy, self).__init__()
        
        
    ########
    ## 
    ## Modify your neural network
    ##
    ########
        
        # 80x80 to outputsize x outputsize
        # outputsize = (inputsize - kernel_size + stride)/stride 
        # (round up if not an integer)
        self.conv1 = nn.Conv2d(2, 4, kernel_size=2, stride=2) # output: 4 matrix of 40x40
        self.conv2 = nn.Conv2d(4, 8, kernel_size=2, stride=2) # output: 8 matrix of 20x20
        self.conv3 = nn.Conv2d(8, 8, kernel_size=2, stride=2) # output: 8 matrix of 10x10
        self.pool = nn.MaxPool2d(2,2) # output: 8 matrix of 5x5
        self.size = 8*5*5
        
        # 1 fully connected layer
        self.fc1 = nn.Linear(self.size, 32)
        self.fc2 = nn.Linear(32, 1)
        self.sig = nn.Sigmoid()
        
        """
        ***Original network***
        self.conv = nn.Conv2d(2, 1, kernel_size=4, stride=4)
        self.size=1*20*20
        
        # 1 fully connected layer
        self.fc1 = nn.Linear(self.size, 1)
        self.sig = nn.Sigmoid()
        """
        
    def forward(self, x):
        
    ########
    ## 
    ## Modify your neural network
    ##
    ########
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = self.pool(x)
        x = x.view(-1,self.size) # flatten the tensor
        x = F.relu(self.fc1(x))
        x = self.sig(self.fc2(x))
        return x
        """
        ***Original Forward function***
        x = F.relu(self.conv(x))
        x = x.view(-1,self.size) # flatten the tensor
        return self.sig(self.fc(x))
        """

# use your own policy!
policy=Policy().to(device)
# policy=pong_utils.Policy().to(device)

# we use the adam optimizer with learning rate 2e-4
# optim.SGD is also possible
import torch.optim as optim
optimizer = optim.Adam(policy.parameters(), lr=1e-4)
```

## Exercise 2: write your own function for training
(this is the same as policy_loss except the negative sign)

### REINFORCE
you have two choices (usually it's useful to divide by the time since we've normalized our rewards and the time of each trajectory is fixed)

1. $\frac{1}{T}\sum^T_t R_{t}^{\rm future}\log(\pi_{\theta'}(a_t|s_t))$
2. $\frac{1}{T}\sum^T_t R_{t}^{\rm future}\frac{\pi_{\theta'}(a_t|s_t)}{\pi_{\theta}(a_t|s_t)}$ where $\theta'=\theta$ and make sure that the no_grad is enabled when performing the division

```python
import numpy as np
def calculate_loss(policy, old_probs, states, actions, rewards,
              gamma = 0.995, beta=0.01):

    ########
    ## 
    ## NOISE REDUCTION --> REWARD NORMALIZATION
    ##
    ########
    
    # get number of trajectories = num of agents
    steps_in_trajectories = len(states) 
    num_trajectories = len(states[0]) 
    
    actions = torch.tensor(actions, dtype=torch.int8, device=device) # ACTIONS: 'RIGHTFIRE' = 4 and 'LEFTFIRE" = 5
    #print('ACTIONS:\n',actions)
    new_probs = pong_utils.states_to_prob(policy, states) # convert states to policy (or probability)
    #print('SATES--> PROB\n',new_probs)
    new_probs = torch.where(actions == pong_utils.RIGHT, new_probs, 1.0-new_probs)
    #print('ONLY RIGHT ACTIONS\n',new_probs)
    
    # REWARD stats
    total_rew_for_each_agent = np.sum(rewards, axis = 0)
    r_mean = np.mean(total_rew_for_each_agent)
    r_std = np.std(total_rew_for_each_agent)
    # get normalized rewards (based on the reward of all trajectories)
    if r_std == 0:
        R_np = np.zeros(num_trajectories)
    else:
        R_np = (total_rew_for_each_agent - r_mean)/r_std 
    
    # POLICY_LOSS
    policy_loss = []
    for i,log_prob in enumerate(new_probs):
        # transform into tensor for computing
        #R = torch.from_numpy(R_np).float() #specify to prepare data for CPU
        R = torch.from_numpy(R_np).float().to(device) #specify to prepare data for CUDA

        # multiply element-wise and store results in list
        result = torch.mul(log_prob,R)
        policy_loss.append(result)
    policy_loss = torch.cat(policy_loss) #concat in single 1D-tensor
    policy_loss = policy_loss.sum(dim = 0) # sum all values
    # calculate the gradient estimation
    policy_loss /= num_trajectories
    
    return policy_loss
    
    # include a regularization term
    # this steers new_policy towards 0.5
    # which prevents policy to become exactly 0 or 1
    # this helps with exploration
    # add in 1.e-10 to avoid log(0) which gives nan
    
    #entropy = -(new_probs*torch.log(old_probs+1.e-10)+ \
    #    (1.0-new_probs)*torch.log(1.0-old_probs+1.e-10))

    #return torch.mean(beta*entropy)


Lsur= calculate_loss(policy, prob, state, action, reward)
print(Lsur)
```

# Training
We are now ready to train our policy!
WARNING: make sure to turn on GPU, which also enables multicore processing. It may take up to 45 minutes even with GPU enabled, otherwise it will take much longer!

```python
from parallelEnv import parallelEnv
import numpy as np
# WARNING: running through all 800 episodes will take 30-45 minutes

# training loop max iterations
episode = 1000
# episode = 800

# widget bar to display progress
!pip install progressbar
import progressbar as pb
widget = ['training loop: ', pb.Percentage(), ' ', 
          pb.Bar(), ' ', pb.ETA() ]
timer = pb.ProgressBar(widgets=widget, maxval=episode).start()

# initialize environment
envs = parallelEnv('PongDeterministic-v4', n=8, seed=1234)

discount_rate = .99
beta = .01
tmax = 100  #320

# keep track of progress
mean_rewards = []

for e in range(episode):

    # collect trajectories
    old_probs, states, actions, rewards = \
        pong_utils.collect_trajectories(envs, policy, tmax=tmax)
        
    total_rewards = np.sum(rewards, axis=0)

    L = -calculate_loss(policy, old_probs, states, actions, rewards, beta=beta) # use your own surrogate function
    # L = -pong_utils.surrogate(policy, old_probs, states, actions, rewards, beta=beta) # this is the SOLUTION!
    optimizer.zero_grad()
    L.backward()
    optimizer.step()
    del L
        
    # the regulation term also reduces
    # this reduces exploration in later runs
    beta*=.995
    
    # get the average reward of the parallel environments
    mean_rewards.append(np.mean(total_rewards))
    
    # display some progress every 20 iterations
    if (e+1)%20 ==0 :
        print("Episode: {0:d}, score: {1:f}".format(e+1,np.mean(total_rewards)))
        print(total_rewards)
        
    # update progress widget bar
    timer.update(e+1)
    
timer.finish()
    
```

```python
# play game after training!
pong_utils.play(env, policy, time=2000) 
```
